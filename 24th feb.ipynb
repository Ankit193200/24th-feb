#  Q2: Write a Python function to re-index the DataFrame.

python
Copy code
import pandas as pd

def reindex_dataframe(df):
    new_index = range(1, len(df) * 2, 2)
    df_reindexed = df.set_index(pd.Index(new_index))
    return df_reindexed

    # Example usage:
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
df_reindexed = reindex_dataframe(df)
print(df_reindexed)

Q7: Write a Python function to extract the 'Username' from the 'Email' column.

Copy code
import pandas as pd

def extract_username(df):
    df['Username'] = df['Email'].apply(lambda x: x.split('@')[0])
    return df

# Example usage:
df = pd.DataFrame({'Email': ['john.doe@example.com', 'alice.smith@example.com']})
df_with_username = extract_username(df)

Q10: Write a Python function to calculate moving average.

python
Copy code
import pandas as pd

def calculate_moving_average(df):
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    return df

# Example usage:
df = pd.DataFrame({'Sales': [10, 20, 30, 40, 50, 60, 70, 80]})
df_with_moving_average = calculate_moving_average(df)
print(df_with_moving_average)
